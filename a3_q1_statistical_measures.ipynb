```python
import pandas as pd
import numpy as np

crime_df = pd.read_csv("crime.csv")

crime_col = crime_df["ViolentCrimesPerPop"]

mean_val = crime_col.mean()
median_val = crime_col.median()
std_val = crime_col.std()
min_val = crime_col.min()
max_val = crime_col.max()


print("Mean:", mean_val)
print("Median:", median_val)
print("Standard Deviation:", std_val)
print("Minimum:", min_val)
print("Maximum:", max_val)
```
