```python
import pandas as pd
import matplotlib.pyplot as plt

crime_df = pd.read_csv("crime.csv")
crime_col = crime_df["ViolentCrimesPerPop"].dropna()

plt.figure()
plt.hist(crime_col, bins=20)
plt.title("Histogram of Violent Crimes Per Population")
plt.xlabel("Violent Crimes Per Population")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.boxplot(crime_col)
plt.title("Boxplot of Violent Crimes Per Population")
plt.ylabel("Violent Crimes Per Population")
plt.show()

```
